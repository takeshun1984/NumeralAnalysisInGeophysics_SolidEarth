$f(x)=e^x\sin{(x)}$の導関数は$f'(x)=e^x(\sin{(x)}+\cos{(x)})$となる。$x=0$のとき、$f'(0)=1$であることを利用して、差分法の精度を数値的に確認する。

このとき、通常格子とスタッガード格子の比較をする。


通常格子
$$
\frac{df(x)}{dx} = \frac{f(x+\Delta x)-f(x-\Delta x)}{2\Delta x}
$$
スタッガード格子
$$
\frac{df(x)}{dx} = \frac{f(x+\Delta x/2)-f(x-\Delta x/2)}{\Delta x}
$$

```python
import numpy as np

def f(x):
    return np.exp(x) * np.sin(x)

# 解析解
true_val = 1.0

# dx を変化させる（小さくしていく）
dx_list = np.logspace(-3, 0, 30)  # 0.01~1を対数スケールで

fd_list  = []
fds_list = []
for dx in dx_list:
    fd  = (f(dx  ) - f(-dx)  ) / (2*dx)
    fds = (f(dx/2) - f(-dx/2)) / (dx)
    fd_list.append(fd)
    fds_list.append(fds)

# 結果表示
for dx, fd, fds in zip(dx_list, fd_list, fds_list):
    print(f"dx={dx:.2e},  FD={fd:.8f},  error={abs(fd - true_val):.2e}, FDS={fds:.8f},  errors={abs(fds - true_val):.2e}")
```

```python
import matplotlib.pyplot as plt

# === プロット ===
plt.rcParams.update({'font.size': 10})
plt.rcParams['xtick.direction'] = 'in' #x軸の目盛りの向き
plt.rcParams['ytick.direction'] = 'in' #y軸の目盛りの向き
plt.rcParams["xtick.minor.visible"] = True  #x軸補助目盛りの追加
plt.rcParams["ytick.minor.visible"] = True  #y軸補助目盛りの追加


fig, axs = plt.subplots(1,2,figsize=(10, 4), sharex=True)


error  = np.abs(np.array(fd_list ) - true_val)
errors = np.abs(np.array(fds_list) - true_val)


ax = axs[0]
ax.semilogx(dx_list, fd_list , marker='o',markersize=5,color='blue',label="Normal FD")
ax.semilogx(dx_list, fds_list, marker='^',markersize=5,color='red' ,label="Staggered FD")
ax.axhline(true_val, linestyle="--", color="gray", label="True value")
ax.set_ylabel("$f'(x=0)$",fontsize=14)
ax.set_ylim(0.8,1.4)
ax.legend(loc='upper left')
ax.set_xlabel(r"$\Delta x$",fontsize=14)

ax = axs[1]
ax.loglog(dx_list, error,  marker='o',markersize=5,color='blue',label="Normal FD")
ax.loglog(dx_list, errors, marker='^',markersize=5,color='red' ,label="Staggered FD")
ax.set_xlabel(r"$\Delta x$",fontsize=14)
ax.set_ylabel("Error",fontsize=14)
ax.set_ylim(1e-7,1)
plt.legend(loc='upper left')
ax.grid(True, which="both")

plt.show()
```
