関数$f(x)$の2次精度中央差分の式は、
$$
\frac{df(x)}{dx} = \frac{f(x+\Delta x)-f(x-\Delta x)}{2\Delta x}
$$
となる。以下、解析的に微分が可能な$f(x)=e^x\sin{(x)}$を用いて、$f'(x)=e^x(\sin{(x)}+\cos{(x)})$と2次精度中央微分による導関数$f'_{FD}(x)$を比較する。

````python
import numpy as np

# x範囲を設定
x = np.linspace(-2, 2, 401)
dx = x[1] - x[0]
print(f"dx : {dx:.4f}")

# 関数
y = np.exp(x) * np.sin(x)

# 理論
dxy = np.exp(x)*(np.sin(x)+np.cos(x)) 

# 中央差分（2次精度）
dxy_fd = np.zeros_like(y)
dxy_fd[1:-1] = (y[2:] - y[:-2]) / (2 * dx)

# for文版
# for i in range(1, n-1):
#    dxy_fd[i] = (y[i+1] - y[i-1]) / (2 * dx)

# 境界は片側差分で代替
# 左端 (index 0)
dxy_fd[0] = (y[1] - y[0]) / dx
# 右端 (index -1)
dxy_fd[-1] = (y[-1] - y[-2]) / dx
````

解析的に微分した結果dxyと、差分を用いた結果dxy_fdを図示して比較してみる。

````python
import matplotlib.pyplot as plt

# === プロット ===
plt.rcParams.update({'font.size': 10}) #全体のフォントサイズ
plt.rcParams['xtick.direction'] = 'in' #x軸の目盛りの向き
plt.rcParams['ytick.direction'] = 'in' #y軸の目盛りの向き
plt.rcParams["xtick.minor.visible"] = True  #x軸補助目盛りの追加
plt.rcParams["ytick.minor.visible"] = True  #y軸補助目盛りの追加


plt.figure(figsize=(4,4))
plt.plot(x, y,color='black',linewidth=1,label='$y = e^x \sin(x)$')
plt.plot(x, dxy,color='lightblue',linewidth=4,label='Analytical')
plt.plot(x, dxy_fd,color='blue',linewidth=1,linestyle='--',label='FD')
plt.xlabel("x")
plt.grid(True)
plt.legend(loc='upper left')
plt.xlim(-2,2)
plt.show()
````

````python
# 相対的な誤差の計算 [%]
diff = (dxy_fd-dxy) / dxy * 100

# === プロット ===
plt.rcParams.update({'font.size': 10}) #全体のフォントサイズ
plt.rcParams['xtick.direction'] = 'in' #x軸の目盛りの向き
plt.rcParams['ytick.direction'] = 'in' #y軸の目盛りの向き
plt.rcParams["xtick.minor.visible"] = True  #x軸補助目盛りの追加
plt.rcParams["ytick.minor.visible"] = True  #y軸補助目盛りの追加


plt.figure(figsize=(4,4))
plt.plot(x, diff,color='blue',linewidth=1,linestyle='-',label='Relative error')
plt.xlabel("x")
plt.ylabel("Relative error [%]")
plt.grid(True)
plt.xlim(-2,2)
plt.tight_layout()
plt.show()
````
