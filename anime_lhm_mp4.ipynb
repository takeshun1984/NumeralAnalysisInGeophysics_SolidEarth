```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import os

# --- 設定 ---
fact = 2e4
dt = 0.02
steps = range(100, 8001, 100)
NX, NZ = 800 // 2, 250 // 2
oname = "lhmPSV.mp4" # 拡張子を mp4 に
xb = np.arange(0, 321, 1)

# 初期プロットの設定
fig, axs = plt.subplots(1, 2, figsize=(15, 3), constrained_layout=True)
plt.rcParams.update({'font.size': 10})

# データの初期化（最初のステップを読み込んで枠を作る）
def load_data(step):
    filename = f"psv.l.{step:05d}.out"
    if not os.path.exists(filename):
        return None
    
    out = pd.read_csv(filename, sep='\s+', names=('X','Z','VX','VZ','div','rot'))
    X = out['X'].values.reshape(NZ, NX)
    Z = out['Z'].values.reshape(NZ, NX)
    VP = out['div'].values.reshape(NZ, NX) * fact * 2
    VS = out['rot'].values.reshape(NZ, NX) * fact
    return X, Z, VP, VS

# 最初のフレームを読み込む
X, Z, VP, VS = load_data(steps[0])

# 左側：div v
im1 = axs[0].pcolormesh(X, Z, VP, shading='auto', cmap='RdBu_r', vmin=-1, vmax=1)
txt1 = axs[0].text(0.95, 0.05, "", transform=axs[0].transAxes, fontsize=14, va='bottom', ha='right')
axs[0].set_xlabel('Distance [km]', fontsize=15)
axs[0].set_ylabel('Depth [km]', fontsize=15)
fig.colorbar(im1, ax=axs[0], shrink=0.4).set_ticks([])

# 右側：rot v
im2 = axs[1].pcolormesh(X, Z, VS, shading='auto', cmap='RdBu_r', vmin=-1, vmax=1)
txt2 = axs[1].text(0.95, 0.05, "", transform=axs[1].transAxes, fontsize=14, va='bottom', ha='right')
axs[1].set_xlabel('Distance [km]', fontsize=15)
fig.colorbar(im2, ax=axs[1], shrink=0.4).set_ticks([])

for ax in axs:
    ax.set_xlim(0, 320)
    ax.set_ylim(90, -10)
    ax.set_aspect(1.0)
    ax.plot(xb, np.zeros_like(xb), ls='-', color='black')
    ax.plot(xb, np.full_like(xb, 3), ls='--', color='black')
    ax.plot(xb, np.full_like(xb, 18), ls='--', color='black')
    ax.plot(xb, np.full_like(xb, 33), ls='--', color='black')
    ax.tick_params(direction="in", top=True, right=True, which='both')

#plt.tight_layout()

# 更新用関数
def update(step):
    data = load_data(step)
    if data is None:
        return im1, im2, txt1, txt2
    
    X, Z, VP, VS = data
    time_val = step * dt
    
    # データの更新（高速）
    im1.set_array(VP.ravel())
    im2.set_array(VS.ravel())
    txt1.set_text(fr"$\mathrm{{div}} \ \mathbf{{v}}$ at {time_val:.2f} s")
    txt2.set_text(fr"$\mathrm{{rot}} \ \mathbf{{v}}$ at {time_val:.2f} s")
    
    print(f"Processing step {step}")
    return im1, im2, txt1, txt2

# アニメーションの作成
ani = FuncAnimation(fig, update, frames=steps, blit=True)

# 保存
print("Saving MP4...")
writer = FFMpegWriter(fps=10, metadata=dict(artist='Me'), bitrate=1800)
ani.save(oname, writer=writer)
plt.close()

print(f"Successfully saved: {oname}")
```
