ここで、波数kとしたときに、
$$
f(n\Delta x) = \exp{(ikn\Delta x)}
$$
とする。すると、この微分は、
$$
\frac{d}{dx}f(n\Delta x)=ik_e \exp{(ikn\Delta x)}
$$
となります。$k_e$は実効的な波数で、解析的には$k_e=k$となります。一方で、2次精度の中心差分で評価しますと、
$$
\frac{d}{dx}f(n\Delta x) \sim \frac{\exp{{ik(n\Delta x+1/2\Delta x)}}-\exp{{ik(n\Delta x-1/2\Delta x)}}}{\Delta x} \\
=i\frac{\sin(k\Delta x/2)}{\Delta x/2}f(n\Delta x)
$$
となります。2次精度の場合、$k_e$は、
$$
k_e = \frac{\sin(k\Delta x/2)}{\Delta x/2}
$$
となります。

```python
import numpy as np
import matplotlib.pyplot as plt

kdx   = np.arange(0.02,4,0.02)
error = 1-np.sin(kdx/2)/(kdx/2) 
```

$k$と$k_e$の誤差を
$$
\frac{k-k_e}{k}
$$
として相対的に評価するとすると、
$$
\frac{k-k_e}{k}=\frac{k-\frac{\sin(k\Delta x/2)}{\Delta x/2}}{k}=1-\frac{\sin{(k\Delta x/2)}}{k\Delta x/2}
$$
となりますので、$k\Delta x$を変数として図を記します。

```python
plt.figure(figsize=(6, 3))

# === プロット ===
plt.rcParams.update({'font.size': 10})
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams["xtick.minor.visible"] = True  
plt.rcParams["ytick.minor.visible"] = True 

plt.plot(kdx,error)
plt.xlabel(r"$k\Delta x$",fontsize=14)
plt.ylabel(r"Error $(k-k_e)/k$",fontsize=14)
plt.xlim(0,3)
plt.ylim(0,0.1)
plt.show()
```

$k\Delta x = 0.3$程度まで誤差が1 %以下となりそうだが、
$$
k\Delta x=\frac{2\pi \Delta x}{\lambda}
$$
となるので、
$$
\frac{\lambda}{\Delta x} = \frac{2\pi}{0.4} \sim 15
$$
逆数をとれば波長あたりの格子数となり、12格子程度必要なこととなる。
